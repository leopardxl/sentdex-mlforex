```julia
dataset = CSV.read("data/GBPUSD1d.txt", header = false)
```

```julia
x   = dataset[1]
bid = dataset[2]
ask = dataset[3]
spread = bid-ask
```

```julia
plot(x, bid)
plot(x, ask)
plot(x, spread, color="green")
```
